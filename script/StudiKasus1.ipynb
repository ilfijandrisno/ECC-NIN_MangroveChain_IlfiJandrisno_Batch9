```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../assets/Result_StudiKasus1.csv")
df.head()
```

```python
df_encoded = pd.get_dummies(df,columns=['permit_status','land_type','boundary_defined'])
df_encoded.head()
```

```python
dataframe = df_encoded.corr()
dataframe
```

```python
plt.figure(figsize=(16, 6))
sns.heatmap(dataframe.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
plt.title("Correlation Heatmap \n Faktor Regulasi vs Metrik Biodiversitas", fontdict={'fontsize':13}, pad=16)
```

```python
# np.triu() fungsi numpy untuk mengisolasi segitiga atas suatu matriks sambil mengubah semua nilai di segitiga bawah menjadi 0
np.triu(np.ones_like(dataframe.corr()))
```

```python
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(dataframe.corr(), dtype=bool))
heatmap = sns.heatmap(dataframe.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
plt.title("Triangle Correlation Heatmap \n Faktor Regulasi vs Metrik Biodiversitas", fontdict={'fontsize':13}, pad=16)
```
